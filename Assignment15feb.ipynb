### Solution (1)

* A multiprocessing is an built-in module in python that allow the system to run multiprocess simultaneously.
* it will enable the breaking of applications into smaller threads that can run independently.

### why it is useful?

* it is useful for cpu bound processes,such as computationally heavy tasks
* it will benefit from having multiple processes

### Solution (2)

#### Multiprocessing:
* Multiprocessing uses two or more cpu's to increas computing power.
* Acquires more memory.
* More efficient as compare to multithreading.
#### Multithreading:
* Multithreading uses single process for multi code segment to increase computing power.
* Requires less memory storage.
* Less efficient as compare to multiprocessing

### Solution (3)

```python
import multiprocessing

def print_func(continent):
    print(f"The name of continent is {continent}.")

if __name__ == "__main__":
    names = ['America', 'Europe', 'Africa']
    procs = []
    proc = multiprocessing.Process(target=print_func,args=(names,))
    procs.append(proc)
    proc.start()

    for proc in procs:
        proc.join()
```

### Solution (4)

* Multiprocssing pool allow you creates multiple python processes and spread across multiple cpu's.
#### Why it is used?
* It is use for parallel execution of function across multiple input values.

### Solution (5)

```python
import multiprocessing
```

```python
def cube(x):
    return x*x*x
if __name__ == "__main__":
    with multiprocessing.Pool(processes=5) as p:
        op = p.map(cube,[1,2,3,4,5,6,3])
        print(op)
```

```python
import multiprocessing
```

```python
def print_number(n):
    i = 0
    i += n
    print(i)
if __name__ == "__main__":
        p1 = multiprocessing.Process(target=print_number,args=(1,))
        p2 = multiprocessing.Process(target=print_number,args=(2,))
        p3 = multiprocessing.Process(target=print_number,args=(3,))
        p4 = multiprocessing.Process(target=print_number,args=(4,))
        p1.start()
        p1.join()
        p2.start()
        p2.join()
        p3.start()
        p3.join()
        p4.start()
        p4.join()
```
